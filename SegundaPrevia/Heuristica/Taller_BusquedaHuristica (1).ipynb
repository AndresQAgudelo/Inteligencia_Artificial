```
graph = {0: [4, 2.0, 6, 1],
         4: [2, 5],
         2.1: [3, 1.0],
         6: [7],
         1: [4.0, 2, 6]
         }




def Minimax (m, profundidad, graph):
  if graph.get(m) == None:
    return m
    
  sucesores = graph.get(m)
  print(m)
  mejor = min(sucesores)

  for n in sucesores:
    M = Minimax(n, profundidad + 1, graph)

    mejor = max(M, mejor)
  
    return mejor

mejor = Minimax(0, 0, graph)
print(mejor)


```
