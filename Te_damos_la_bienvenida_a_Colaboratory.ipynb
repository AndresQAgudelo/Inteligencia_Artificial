```
graph = {0: [4, 2, 6, 1],
         4: [2, 5],
         2: [3, 1],
         6: [7],
         1: [4, 2, 6]
         }




def MiniMax(source , graph):
  
  stack = graph.get(source)

  while True:
    mejormax = max(stack)

    stack2 = graph.get(mejormax)

    mejormin = min(stack2)

    mejor = mejormax - mejormin

    if mejor >= 1:
      print (f"Mejor jugada es :  {mejormax} - {mejormin}")
      break
    
    else:
      stack.remove(mejormax)

MiniMax(0, graph)

```
